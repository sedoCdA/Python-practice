````python
# Outlier detection (IQR method)
Q1 = df['sepal length (cm)'].quantile(0.25)
Q3 = df['sepal length (cm)'].quantile(0.75)
IQR = Q3-Q1
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR

print("Upper Bound : ", upper_bound)
print("Lower Bound : ", lower_bound)
````

````python
my_dict = {
    "income":(50, 25, 15, 10, np.nan, np.nan, 20, 10, np.nan, 999),
    "age":(19, 20, 30, np.nan, np.nan, 25, 20, 52, 46, 9999)
}
my_dict
````

````python
data = pd.DataFrame(my_dict)
data
````

````python
data = data.replace((999, 9999),np.nan)
data
````

````python
# Display values with null as True
data.isnull()
````

````python
new = data.dropna(how='any')#Removes rows where ever null is there
new
````

````python
new = data.dropna(how='all')#Removes rows where both colums null
new
````

````python
#Replace missing values with mean of respective column
new = np.round(data.fillna(data.mean()),2)
new
````

````python
#Replace missing values with median of respective column
new = np.round(data.fillna(data.median()),2)
new
````
